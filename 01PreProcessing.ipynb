```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = data.select_dtypes(include="number").columns

plt.figure(figsize=(15, 10))

for i, col in enumerate(num_cols):
    plt.subplot(3, 3, i+1)
    sns.boxplot(y=data[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```
